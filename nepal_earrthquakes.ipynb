```python
# Q1 =
# Q3 =
# IQR =
# upper_fence =
# lower_fence =

out_list = ['latitude', 'longitude', 'depth', 'mag', 'nst', 'gap', 'rms', 'horizontalError', 'depthError']

def outliers(cols):
  outliers_list = { }
  for i in cols:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    upper_fence = Q3 + 1.5*IQR
    lower_fence = Q1 - 1.5 * IQR
    outliers = df[(df[i] > upper_fence) | (df[i] < lower_fence)]
    outliers_list[i]=(len(outliers))
  return outliers_list

outliers(out_list)
```
